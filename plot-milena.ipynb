```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
SCALE = 1
matrix_df = pd.read_csv('milena/PlacasPonta.csv', index_col=0, sep=';')
```

```python
normalized = matrix_df.reset_index()
```

```python
pivot_normalized = normalized.pivot(index='x', columns='y', values='tensao')
```

```python
gradient = np.gradient(pivot_normalized)
```

```python
sb.heatmap(pivot_normalized / SCALE)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(pivot_normalized / SCALE, cmap='magma')
ax.clabel(CS, inline=1, fontsize=10)

ax.quiver(-gradient[1], -gradient[0], units='xy', color='gray')

ax.set_title('Gráfico 2: curvas equipotenciais sob efeito de “ponta”')
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
plt.show()
```
